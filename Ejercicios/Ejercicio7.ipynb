cajeras van a ir atendiendo a los clientes  


Si cliente es VIP, 15% en la compra

```python
class Tvision:
    
    def __init__(self, brand, ide, inches, price):
        self.brand = brand
        self.ide = ide
        self.inches = inches
        self.price = price
        print("Television creada con exito")
        
    def __str__(self):
        return "television de la marca " + self.brand
        
class Monitor:
    
    def __init__(self, brand, ide, inches, price):
        self.brand = brand
        self.ide = ide
        self.inches = inches
        self.price = price
        print("monitor creado con exito")
        
    def __str__(self):
        return "monitor de la marca " + self.brand
        
class Computer:
    
    def __init__(self, brand, ide, hard_disk, ram, price):
        self.brand = brand
        self.ide = ide
        self.hard_disk = hard_disk
        self.ram = ram
        self.price = price
        print("ordenador creado con exito")
        
    def __str__(self):
        return "ordenador de la marca " + self.brand

class Client:
    shopping_cart = []
    
    def __init__(self, name, vip):
        self.name = name
        self.vip = vip
        print("cliente creado con exito")
        
    def pay(self, total_price):
        print("el cliente ha pagado: ", str(total_price))
        
class Cashier:
    
    def __init__(self, name, number):
        self.name = name
        self.number = number
        print("cajera creada con exito")
    
    def charge(self, client):
        total_price = 0
        for product in client.shopping_cart:
            total_price += product.price
            
        if client.vip:
            discount = (total_price*15)/100
            total_price = total_price - discount
            print("la cajera {} ha cobrado: {} aplicando el descuento vip".format(self.name, total_price))
        else:
            print("la cajera {} ha cobrado: {} ".format(self.name, total_price))
        
        client.pay(total_price)

class Shop:

    def __init__(self, clients, cashiers):
        self.clients = clients
        self.cashiers = cashiers
        print("tienda creada con exito")
        
    def buy(self):
        for client in self.clients:
            cart = list(map(lambda product: str(product), client.shopping_cart))
            print("el cliente {} se dispone a comprar: {}".format(client.name, cart))
            if self.clients.index(client)%2 == 0:
                print("el cliente {} es atendido por la cajera {}".format(client.name, self.cashiers[0].name))
                self.cashiers[0].charge(client)
            else:
                print("el cliente {} es atendido por la cajera {}".format(client.name, self.cashiers[1].name))
                self.cashiers[1].charge(client)
```

```python
tv = Tvision("LG", "1", "30", 300)
monitor = Monitor("Samsung", "2", "50", 800)
computer = Computer("HP", "3", "500GB", "16GB", 1000)
client1 = Client("Marco", False)
client2 = Client("Edward", True)
client3 = Client("Ace", False)
cashier1 = Cashier("Vivi", "1234")
cashier2 = Cashier("Bonney", "5678")
shop = Shop([client1, client2, client3], [cashier1, cashier2])
client1.shopping_cart = [computer]
client2.shopping_cart = [computer, tv]
client3.shopping_cart = [computer, monitor]
shop.buy()
```
